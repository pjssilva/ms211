# Problemas de Valor de Contorno

Estudamos anteriormente problemas de valor inicial, onde toda a dinâmica está definida pela equação diferencial e os valores da função desconhecida, e eventualmente de algumas de suas derivadas, em um instante inicial $t_0$. Em particular, se a equação diferencial envolve $y$, $y'$ e $y''$ consideramos um problema na forma
\begin{align*}
y''(t)  &= f(t, y(t), y'(t))\quad\quad t > t_0, \\
y(t_0)  &= \alpha, \\
y'(t_0) &= \beta.
\end{align*}

Esse tipo de problema é bem natural se estamos estudando a evolução de algo no tempo, já que nesse caso é natural que a situação do objeto estudado seja conhecida no tempo inicial $t_0$. Note que as condições iniciais são chave para garantir que a solução do sistema seja única.

Porém há situações em que não conhecemos o valor da derivada de $y$ no ponto inicial. Se tivermos apenas o valor de $y$ em um ponto ainda sobrará um grau de liberdade e a equação diferencial não terá a sua solução completamente determinada. Para evitar isso uma outroa possibilidade é conhecermos o valor de $y$ em um outro ponto além do inicial. Por exemplo, podemos estar interesados em considerar problemas do tipo
\begin{align*}
y''(x)  &= f(x, y(x), y'(x))\quad\quad a < x < b, \\
y(a) &= \alpha, \\
y(b) &= \beta.
\end{align*}
Nesse caso conhecemos os valores que de $y$ em dois pontos, que chamamos de *valores de contorno*. Um exemplo é por exemplo estudar o movimento de um projétil quando conhecemos apenas a força que age sobre ele e a sua posição em dois intantes de tempo mas não conhecemos a sua velocidade inicial. 

Nessa parte do curso vamos estudar métodos numéricos para resolução do problema acima que é conhecido como *problema de valores de contorno (PVC)*.

## Método das diferenças finitas

O método da diferença finitas começa sendo definido como os métodos para problemas de valor inicial. Inicialmente vamos desistir de encontrar uma função como solução. Aceitamos apenas considerar uma discretização do intervalo $a = x_0 < x_1 < x_2 < \ldots < x_N = b$ e o nosso objetivo será simplesmente encontrar aproximações $y_i$ do valores da função solução nos pontos $t_i$. Ou seja queremos encontrar $y_i$ tal que
$$
y_i\approx y(t_i),\quad\quad i = 0, \ldots, N,
$$
em que $y$ é a solução do PVC.

A grande dificuldade de resolve a equação diferencial é que ela envolve os valores das derivadas (primeira e segunda) da função icógnita. Vamos tentar evitar isso aproximando essas dererivadas por fórmulas que dependem apenas dos valores da função. A melhor maneira de conseguir isso é olhar para as expansões de Taylor de $y$. Recapitulando, se $y$ é $n + 1$ diferenciável temos
$$
y(x + h) = y(x) + h y'(x) + \frac{h^2}{2} y''(x) + \ldots + \frac{h^n}{n!} y^{(n)}(x) + \frac{h^{n + 1}}{(n + 1)!} y^{(n + 1)}(\xi),\quad \xi \in (x, x + h).
$$
Lembrem mais uma vez que o ponto $\xi$ que aparece na fórmula acima não é conhecido exatamente, apenas sabemos que ele pertence ao intervalo $(x, x + h)$. 

Usando essa fórmula é possível obter aproximaçòes para a primeira derivada de $y$ usando apenas os valores da função.
$$
y(x_{i + 1}) = y(x_i) + h y'(x_i) + \frac{h^2}{2} y''(\xi).
$$
Portanto
$$
y'(x_i) = \frac{y(x_{i + 1}) - y(x_i)}{h} - \frac{h}{2} y''(\xi).
$$
E, se $h$ é pequeno, podemos escrever
$$
y'(x_i) \approx \frac{y(x_{i + 1}) - y(x_i)}{h}.
$$
Essa fórmula é conhecida como *diferença avançada* e vemos que o erro nela é proporcial a $h$, que vamos escrever é da ordem $h$ e usar a notação $O(h)$.

De maneira análoga podemos deduzir a fórmula de *diferença atrasada*: 
$$
y'(x_i) \approx \frac{y(x_{i}) - y(x_{i - 1})}{h}.
$$
De novo uma fórmula com erro da ordem de $h$. 

Mais interessante ainda podemos aproveitar as duas fórmulas de Taylor usadas acima e conseguir algo ainda melhor
\begin{align*}
y(x_{i + 1}) &= y(x_i) + h y'(x_i) + \frac{h^2}{2} y''(x_i) + \frac{h^3}{3!} y'''(\xi^+),\quad \xi^+ \in (x_i, x_{i + 1}) \\
y(x_{i - 1}) &= y(x_i) - h y'(x_i) + \frac{h^2}{2} y''(x_i) - \frac{h^3}{3!} y'''(\xi^-),\quad \xi^- \in (x_{i - 1}, x_i).
\end{align*}
De forma análoga ao que foi feito no método do ponto médio na seção anterior, ao subtrairmos essas duas fórmulas obtemos
$$
y(x_{i + 1}) - y(x_{i -1}) = 2hy' (x_i) + O(h^3).
$$
De onde obtemos
$$
y' (x_ i) = \frac{y(x_{i + 1}) - y(x_{i -1})}{2h} + O(h^2).
$$
Essa é uma aproximação de ordem melhor para $y'$ do que as duas anteriores já que o erro dela ela é proporcional a $h^2$ que pode ser bem menor que $h$. Essa fórmula é conhecida como *diferença centrada* e é geral preferida às diferenças avançada e atrasada.

Agora para podermos atacar o PVC como descrito acima, que envolve também as dereivadas segundas de $y$, precisamos de alguma aproximação para essa derivada. De novo partimos das expansões de Taylor de $x_{i - 1}$ e $x_{i + 1}$ em torno de $x_i$.
\begin{align*}
y(x_{i + 1}) &= y(x_i) + h y'(x_i) + \frac{h^2}{2} y''(x_i) + \frac{h^3}{3!} y'''(\xi) + \frac{h^4}{4!} y(\xi^+),\quad \xi^+ \in (x_i, x_{i + 1}) \\
y(x_{i - 1}) &= y(x_i) - h y'(x_i) + \frac{h^2}{2} y''(x_i) - \frac{h^3}{3!} y'''(\xi) + \frac{h^4}{4!} y(\xi^-), \quad \xi^- \in (x_{i - 1}, x_i).
\end{align*}
Só que agora ao invés de subtrair vamos somar as expressões para obter
$$
y(x_{i + 1}) + y(x_{i - 1}) = 2y(x_i) + h^2 y''(x_i) + O(h^4).
$$
Isolando o $y''(x_i)$ que é o que desejamos aproximar obtemos
$$
y''(x_i) = \frac{y(x_{i + 1}) - 2y(x_ i) + y(x_{i - 1})}{h^2} + O(h^2).
$$
Isso quer dizer que a aproximação para a derivada
$$
y''(x_i) \approx \frac{y(x_{i + 1}) - 2y(x_ i) + y(x_{i - 1})}{h^2},
$$
também é da ordem de $h^2$, assim como a diferença centrada para aproximar $y'$. Essa aproximação é conhecida també como *diferença centrada*, mas agora para aproximar a segunda derivada.

Estamos prontos para colocar isso em prática para resolver um PVC.

## Problemas lineares

Vamos começar pensando como resolver um problema de valores de contorno linear. Nesse caso consideramos que a função $f$ depente de forma linear de $y$ e $y'$. Para ser mais preciso, vamos considerar que 
$$
f(x, y, y') = -p(x)y' - q(x)y + r(x),
$$
para funções $p,\ q,\ r$ que dependem apenas de $x$. Assim o PVC fica
$$
y'' + p(x)y' + q(x)y = r(x),\quad\quad y(a) = \alpha,\quad y(b) = \beta.
$$

### Exemplo
$$
y'' - 5xy' + x^2y = e^x,\quad\quad y(0) = 3,\quad y(1) = 0.
$$
Nesse caso identificamos as funções $p(x) = -5x$, $q(x) = x^2$ e $r(x) = e^x$. 

## O Método de diferenças finitas no exemplo

O método de diferenças finitas consiste em usar a equação diferencial e aproximações de derivadas calculadas anteriormente para definir equações que definiem as aproximações $y_i \approx y(x_i)$. Por exemplo, se considerarmos o exemplo anterior e os pontos $0 = x_0 < x_1 = 1/4 < x_2 = 1/2 < x_3 = 3/4 < x_4 = 1$ podemos escrever as condições dadas pela equação diferencial como 
\begin{align*}
&y''(x_i) - 5 x_i y'(x_i) + x_i^2y(x_i) = e^{x_i},\quad\quad i = 1,\ldots,3, \\
&y(x_0) = 3, \\
&y(x_4) = 0,
\end{align*}
Agora substituindo os valores exatos $y(t_i)$ por suas aproximações $y_i$, as condições de contorno e as aproximações de derivadas centradas teremos
\begin{align*}
&\frac{y_{i + 1} - 2y_i + y_{i - 1}}{h^2} - 5 x_i \frac{y_{i + 1} - y_{i -1}}{2h}  + x_i^2 y_i = e^{x^i}, \quad\quad i = 1,\ldots,3, \\
&y_0 = 3, \\
&y_4 = 0.
\end{align*}
Organizando as equações acima multiplicadas por $h ^2$ temos
\begin{align*}
&(1 - (h/2) 5x_i)y_{i - 1} - (2 + h ^2 x_i^2)y_i + (1 + (h/2) 5x_i)y_{i + 1} = h^2 e^{x_i},\quad\quad i = 1,\ldots,3, \\
&y_0 = 3, \\
&y_4 = 0.
\end{align*}
Como os valores de $h$ e $x_i, i = 0,\ldots,4$, são conhecidos o que temos acima um sistema de equações lineares nas variáveis $y_1, y_2, y_3$. Lembre-se que $y_0$ e $y_4$ já tem os seus valores definidos. Ou seja um sistema simples $3$ por $3$.

Note que aqui não é possível calcular $y_1$ somente a partir de $y_0$ como feito em problemas de valor inicial. Isso ocorre porque o valor em $y_1$ depende tanto da condição em $a = x_0$ quando da condição no outro extremo $b = x_4$. Todos as aproximações de $y$ que são desconhecidas devem ser calculadas de uma única vez resolvendo um sistema de equações.

## Fórmula geral do método de diferenças finitas para problemas lineares

Deve ser natural imaginar que podemos escrever uma fórmula geral, que pode então ser implementada como um programa de computador, para a aplicação do método de diferenças finitas no caso linear geral. Vejamos como fazer isso. Vamos iniciar revistando a fórmula do caso geral.
$$
y'' + p(x)y' + q(x)y = r(x),\quad\quad y(a) = \alpha,\quad y(b) = \beta.
$$

A ideia do método de fierenças finitas é dividir o intervalo de solução em subintervalos usando um passo fixo $h$ de modo que $a = x_0 < x_1 < \ldots < x_n = b$ de modo que $x_{i + 1} = x_{i} + h$. A partir daí devemos re-escrever a equação diferencial nos pontos da malha, isto é:
$$
y''(x_i) + p(x_i)y'(x_i) + q(x_i)y(x_i) = r(x_i),\quad i = 1, 2, \ldots, n - 1.
$$

Por fim, devemos substituir as derivadas de $y$ pelas aproximações baseadas em diferenças. Também é natural usar a melhor aproximação que temos isso é, no caso das derivadas primeiras devemos usar diferenças centradas. Ficamos com
\begin{gather}
\frac{y_{i + 1} - 2 y_i + y_{i - 1}}{h^2} + p_i \frac{y_{i+1} - y_{i - 1}}{2h} + q_i y_i = r_i,\quad 
i = 1, 2, \ldots, n - 1. \\
y_0 = \alpha, \quad\quad y_{n + 1} = \beta.
\end{gather}
Também é comum multiplicar as equações acima por $h^2$ para eliminar as frações. Ficamos finalmente com
\begin{gather}
(y_{i + 1} - 2 y_i + y_{i - 1}) + p_i h \frac{y_{i+1} - y_{i - 1}}{2} + h^2 q_i y_i = h^2 r_i,\quad 
i = 1, 2, \ldots, n - 1. \\
y_0 = \alpha, \quad\quad y_{n + 1} = \beta.
\end{gather}
Separando os termos que aparecem na frente das variáveis $y_{i - 1}, y_i, y_{i + 1}$ chegamos a
\begin{gather}
(1 - h/2\ p_i) y_{i - 1} + (-2 + h^2 q_i) y_i + (1 + h/2\ p_i) y_{i + 1} = h^2 r_i,
i = 1, 2, \ldots, n - 1. \\
y_0 = \alpha, \quad\quad y_{n + 1} = \beta.
\end{gather}
Um sistema de equações lineares.

Notem que apesar do sistema acima parecer ser inicialmente de $n + 1$ equações e variáveis, duas variáveis, $y_0$ e$y_n$, tem os seus valores definidos trivialmente pelas duas últimas equações e portanto sobra apenas um sitema nas $n - 1$ variáveis $y_1, y_2, \ldots, y_{n - 1}$.

Podemos também apresentar o sistema em seu formato matricial. Nesse caso ele fica
$$
A y = b.
$$
Com 
$$
A = \left( \begin{array}{ccccc}
b_1 & c_1 & & & \\
a_2 & b_2 & c_2 & & \\
& \ddots & \ddots & \ddots & \\
& & a_{n - 2} & b_{n - 2} & c_{n - 2} \\
& & & a_{n - 1} & b_{n - 1}
\end{array} \right)
\quad\quad
b = \left( \begin{array}{c}
h^2 r_1 - a_1 \alpha \\
h^2 r_2 \\
\vdots \\
h^2 r_{n - 2} \\
h^2 r_{n - 1} - c_{n - 1} y_n
\end{array} \right).
$$
$$
a_i = 1 - h/2\ p_i,\quad\quad b_i = -2 + h^2q_i,\quad\quad c_i = 1 + h/2\ p_i,\quad\quad i = 1, 2, \ldots, n - 1.
$$

Uma observação importante é que a matriz do sistema é tridiagonal. Ela possui apenas a diagonal principal e as primeiras diagonis superior e inferior não nulas. Isso permite resolver o sistema linear muito mais rapidamente, já que, por exemplo, a fatoração LU nesse caso tem que zerar apenas um número por coluna. Uma implementação que leva isso em consideração tanto no processo de fatoração quanto no processo de resolução dos subproblemas por substituição é então capaz de resolver esse problema com grande eficiência, em $O(n)$ ao invés da ordem cúbica que está usualmente associada a fatorações.

Além disso a existência das fórmulas acima mostram que é possível escrever um programa de computador que resolve um PVC. O que esse programa deve fazer é montar a matriz e o lado direito descritos acima e resolver o sistema de forma eficiente.
